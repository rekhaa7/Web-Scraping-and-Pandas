```python
url = 'https://www.worldometers.info/coronavirus/'
```

```python
import requests
```

```python
content = requests.get(url).content
```

```python
type(content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(content, 'html.parser')  #lxml is a parsing tool that converts from one state to another
```

```python
soup
```

```python
tbody = soup.find('tbody')
```

```python
tbody
```

```python
table_row = tbody.find_all('tr')
```

```python
table_row
```

```python
# Garbage data upto 6 index and starts actual data from 8 in Worldometer url

table_row[8]
```

```python
table_row[6]
```

```python
all_country_data = list()
for i in range(8, len(table_row)):
    country = []
    country_data_row = table_row[i].find_all('td')
    for tag in country_data_row:
        country.append(tag.text)
    all_country_data.append(country)
```

```python
all_country_data[0]
```

```python
data = list(map(lambda x: x[1:7], all_country_data))
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data) # DataFrame is a 2-dimensional labeled data structure with columns of potentially different types.
```

```python
df.head()
```

```python
columns = ['Country',
          'Total Cases',
          'New Cases',
          'Total Deaths',
          'New Deaths',
          'Total Recovered']
```

```python
df.columns= columns
```

```python
df.head()
```

```python
df.head(7)
```

```python
# converting dataframe to csv file
df.to_csv('covid-mar-20.csv',index = False) #csv=comma separated value
```
